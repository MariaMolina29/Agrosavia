```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import IFrame

# 1. Función para ejecutar FastQC y MultiQC
def run_fastqc(input_files, output_dir):
    """Ejecuta FastQC y MultiQC para verificar la calidad de las lecturas"""
    os.makedirs(output_dir, exist_ok=True)
    # Ejecución de FastQC
    os.system(f"fastqc -o {output_dir} {' '.join(input_files)}")
    # Ejecución de MultiQC
    os.system(f"multiqc {output_dir} -o {output_dir}/multiqc_output")
    # Visualización de MultiQC
    #IFrame(f"{output_dir}/multiqc_output/multiqc_report.html", width=1000, height=1000)

# 2. Función para preprocesar las secuencias con Trimmomatic y Trim Galore
def trim_sequences(input_files, output_dir):
    """Recorta adaptadores y secuencias de baja calidad"""
    # Trimmomatic
    os.makedirs(output_dir, exist_ok=True)
    os.system(f"trimmomatic PE -threads 4 {input_files[0]} {input_files[1]} "
              f"{output_dir}/1_paired.fq {output_dir}/1_unpaired.fq "
              f"{output_dir}/2_paired.fq {output_dir}/2_unpaired.fq "
              "ILLUMINACLIP:TruSeq3-PE-2.fa:2:30:7 LEADING:3 TRAILING:10 "
              "SLIDINGWINDOW:4:18 MINLEN:50")
    # Trim Galore
    os.system(f"trim_galore --paired --clip_R1 10 --clip_R2 10 --three_prime_clip_R1 10 --three_prime_clip_R2 10 "
          f"--quality 20 --length 50 --fastqc -o {output_dir}/fastqc --basename trimgalore "
          f"{input_files[0]} {input_files[1]}")


# 3. Función para eliminar colas PolyG con Fastp
def trim_polyG(input_files, output_dir):
    """Elimina colas PolyG con Fastp"""
    os.makedirs(output_dir, exist_ok=True)
    os.system(f"fastp -i {input_files[0]} -I {input_files[1]} "
              f"-o {output_dir}/1_polyG.fq -O {output_dir}/2_polyG.fq --trim_poly_g")

# 4. Función para corregir errores con Lighter
def correct_errors(input_files, output_dir):
    """Corrige errores usando Lighter"""
    os.makedirs(output_dir, exist_ok=True)
    os.system(f"lighter -r {input_files[0]} -r {input_files[1]} -k 19 -maxcor 2 -od {output_dir}/lighter_out -t 4")

# 5. Función para ejecutar Kraken 2 y generar el reporte
def run_kraken(input_files, db_path, output_dir):
    """Ejecuta Kraken 2 para la clasificación taxonómica"""
    os.makedirs(output_dir, exist_ok=True)
    os.system(f"kraken2 --db {db_path} --threads 4 --output {output_dir}/kraken_output.txt "
              f"--report {output_dir}/kraken_report.txt {' '.join(input_files)}")

# 6. Función para leer y visualizar el reporte de Kraken
def visualize_kraken_report(report_path):
    """Lee el reporte de Kraken 2 y lo visualiza en un gráfico"""
    kraken_report = pd.read_csv(report_path, sep='\t', header=None, 
                                names=["Percent", "TotalReads", "AssignedReads", "TaxonLevel", "TaxonID", "TaxonName"])
    display(kraken_report.head())
    # Filtrar solo los taxones relevantes
    kraken_report_filtered = kraken_report[kraken_report["TaxonLevel"].isin(["R", "R1", "R2", "K"])]
    display(kraken_report_filtered.head()) 
    
    # Gráfico de barras
    plt.figure(figsize=(10, 6))
    plt.barh(kraken_report_filtered['TaxonName'], kraken_report_filtered['AssignedReads'], color='skyblue')
    plt.xlabel('Número de lecturas asignadas')
    plt.ylabel('Taxón')
    plt.title('Lecturas asignadas por taxón')
    plt.gca().invert_yaxis()  # Invertir el eje Y para mostrar el mayor en la parte superior
    plt.show()

# ------------------------------------------
# Ejemplo de uso con una secuencia específica

# Definir archivos de entrada y directorio de salida
input_files = ["B028_24/BO28_24_1_subset.fastq", "B028_24/BO28_24_2_subset.fastq"]
output_dir = "B028_24/sub"

# 1. Verificar calidad con FastQC y MultiQC
run_fastqc(input_files, output_dir)

# 2. Preprocesar las secuencias con Trimmomatic y Trim Galore
trim_sequences(input_files, output_dir)

# 3. Eliminar colas PolyG con Fast
trim_polyG([f"{output_dir}/fastqc/trimgalore_val_1.fq", f"{output_dir}/fastqc/trimgalore_val_2.fq"], output_dir)

# 4. Corregir errores con Lighter
correct_errors([f"{output_dir}/1_polyG.fq", f"{output_dir}/2_polyG.fq"], output_dir)

# 5. Verificar calidad post-corrección
run_fastqc([f"{output_dir}/lighter_out/1_polyG.cor.fq", 
            f"{output_dir}/lighter_out/2_polyG.cor.fq"], output_dir)

# 6. Ejecutar Kraken 2 para clasificación taxonómica
run_kraken([f"{output_dir}/lighter_out/1_polyG.cor.fq",
            f"{output_dir}/lighter_out/2_polyG.cor.fq"], 
           "~/k2_standard", output_dir)

# 7. Visualizar los resultados de Kraken
visualize_kraken_report(f"{output_dir}/kraken_report.txt")
```
